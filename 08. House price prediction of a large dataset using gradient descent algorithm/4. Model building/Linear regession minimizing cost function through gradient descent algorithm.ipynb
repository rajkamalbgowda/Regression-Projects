## Normalizing the variables

```python
def featureNormalization(X):
    
    mean=np.mean(X,axis=0)
    std=np.std(X,axis=0)
    
    X_norm = (X - mean)/std
    
    return X_norm , mean , std
```

```python
X_normalized, mean_X, std_X=featureNormalization(X_matrix)
```

## Computing cost function

```python
def computeCost(X, y,theta):
    
    m = y.size
    J = 0
    h = X.dot(theta)
    J = 1/(2*m)*(np.sum(np.square(h-y)))
    return(J)
```

```python
computeCost(X_ones,y_vector,theta)
```

```python
Gradient Descent to minimize the cost function
```

```python
def gradientDescent(X, y, theta, alpha=0.01, iterations=1500):
    m = y.size
    J_old = np.zeros(iterations)
    
    for num in np.arange(iterations):
        h = X.dot(theta)
        theta = theta - alpha*(1/m)*(X.T.dot(h-y))
        J_old[num] = computeCost(X, y, theta)
    return(theta, J_old)
```

```python
theta,J_old=gradientDescent(X_ones,y_vector,theta)
```

```python
np.array(J_old).min() #Minimized cost function
```

```python
plt.plot(J_old)

plt.xlabel("Iteration")
plt.ylabel("$J(\Theta)$")

plt.title("Cost function using Gradient Descent")
```
